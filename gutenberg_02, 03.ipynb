## (1) austen - emma, persuasion, sense

```python
from nltk.corpus import gutenberg
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk
from nltk.tokenize import RegexpTokenizer
from collections import Counter
```

```python
import numpy as np
import re
from nltk.corpus import stopwords

emma = gutenberg.open('austen-emma.txt')
emma = emma.readlines()
emma = list(filter(None, [item.strip('\n') for item in emma]))
emma_tokens = [sentence.split() for sentence in emma] 
emma_words = [word for sentence in emma_tokens for word in sentence] 
emma_words = list(filter(None, emma_words))
emma_words = [word.lower() for word in emma_words]
emma_dist= Counter(emma_words)
print('가장 자주 사용된 단어: ',emma_dist.most_common(20))
print()

stopwords = stopwords.words('english')
emma_words = [word for word in emma_words if word not in stopwords]
print('불용어 제거 후: ',Counter(emma_words).most_common(20))
print()

emma = Counter(emma_words).most_common(20)
dict_emma=dict(emma)
plt.pie(dict_emma.values(), labels=dict_emma.keys(), autopct='%.1f%%', pctdistance=0.85, radius=1.5)
plt.title("austen-emma", position=(0.5, 1.2))
plt.show()

token_len = []           # 문장 당 토큰 수
for token in emma_words:
    token_len.append(len(token))
print("token_len:", token_len[:20])

%matplotlib inline

import matplotlib.pyplot
import seaborn

seaborn.distplot(token_len)
   
```
